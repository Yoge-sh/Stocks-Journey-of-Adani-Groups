## <p style="background-color:#17BECF;font-family:newtimeroman;color:#rgb(102,17,0);font-size:150%;text-align:center;border-style: solid;border-color: black;border-radius:10px">Converting Dataframes into Time Series</p>

```python
for df in df_names:
    df.set_index("Date", inplace=True)
df.head(3)
```

```python
plt.figure(figsize=(10,6))
plt.style.use('fivethirtyeight')
df.Close_Price.plot();
```

```python
def volume_shocks(stock):
    """
    'Volume' - Vol_t
    'Volume next day - vol_t+1
    
    """
    stock["vol_t+1"] = stock.Volume.shift(1)  #next rows value
    
    stock["volume_shock"] = ((abs(stock["vol_t+1"] - stock["Volume"])/stock["Volume"]*100)  > 10).astype(int)
    
    return stock
```

```python
for df in df_names:
    volume_shocks(df)
```

```python
# Volume Shock Direction
def direction_fun(stock):
    # considerng only shock - 1 valued rows.
    # 0 - negative and 1- positive
    if stock["volume_shock"] == 0:
        pass
    else:
        if (stock["vol_t+1"] - stock["Volume"]) < 0:
            return 0
        else:
            return 1
```

```python
def vol_shock_direction(stock):
    stock['VOL_SHOCK_DIR'] = 'Nan'
    stock['VOL_SHOCK_DIR'] = stock.apply(direction_fun, axis=1)
    return stock
```

```python
def vol_shock_direction(stock):
    stock['VOL_SHOCK_DIR'] = 'Nan'
    stock['VOL_SHOCK_DIR'] = stock.apply(direction_fun, axis=1)
    return stock
```

```python
for df in df_names:
    vol_shock_direction(df)
```

```python
def price_shocks(stock):
    """
    'ClosePrice' - Close_t
    'Close Price next day - vol_t+1
    
    """
    stock["price_t+1"] = stock.Close_Price.shift(1)  #next rows value
    
    stock["price_shock"] = (abs((stock["price_t+1"] - stock["Close_Price"])/stock["Close_Price"]*100)  > 2).astype(int)
    
    stock["price_black_swan"] = stock['price_shock'] # Since both had same data anad info/
    
    return stock
```

```python
for df in df_names:
    price_shocks(df)
```

```python
def direction_fun_price(stock):
    # considerng only shock - 1 valued rows.
    # 0 - negative and 1- positive
    if stock["price_shock"] == 0:
        pass
    else:
        if (stock["price_t+1"] - stock["Close_Price"]) < 0:
            return 0
        else:
            return 1
```

```python
def price_shock_direction(stock):
    stock['PRICE_SHOCK_DIR'] = 'Nan'
    stock['PRICE_SHOCK_DIR'] = stock.apply(direction_fun_price, axis=1)
    return stock
```

```python
for df in df_names:
    price_shock_direction(df)
```

```python
def price_shock_wo_vol_shock(stock):
    
    stock["not_vol_shock"]  = (~(stock["volume_shock"].astype(bool))).astype(int)
    stock["price_shock_w/0_vol_shock"] = stock["not_vol_shock"] & stock["price_shock"]
    
    return stock
```

```python
for df in df_names:
    price_shock_wo_vol_shock(df)
```

# <p style="background-color:#17BECF;font-family:newtimeroman;color:#rgb(102,17,0);font-size:200%;text-align:center;border-style: solid;border-color: black;border-radius:10px">Bokeh Visualizations</p>

```python
import bokeh
from bokeh.plotting import figure, output_file, show
from bokeh.io import show, output_notebook
from bokeh.palettes import Blues9
from bokeh.palettes import RdBu3
from bokeh.models import ColumnDataSource, CategoricalColorMapper, ContinuousColorMapper
from bokeh.palettes import Spectral11
```

```python
output_notebook()
```

```python
def bokeh_plot(stock,name):
    data = dict(stock=stock['Close_Price'], Date=stock.index)
    
    p = figure(plot_width=800, plot_height=250,  title = 'time series for {}'.format(name), x_axis_type="datetime")
    p.line(stock.index, stock['Close_Price'], color='blue', alpha=0.5)
    
    #show price shock w/o vol shock
    
    p.circle(stock.index, stock.Close_Price*stock["price_shock_w/0_vol_shock"], size=4, legend_label='price shock without volume shock')
    show(p)
```

```python
output_file("timeseries.html")
for df,name in zip(df_names,names):
    bokeh_plot(df,name)
```
